```python
x, s, d, t, y = symbols("x s d t y")
y = Function('y')
f = y(s)**3 - y(s)**2
y0 = d

def phi(k):
    if k == 0:
        return d
    else:
        return y0 + integrate(phi(k-1)**2 - phi(k-1)**3,(s,0,2/d))

phi(1).simplify()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def phi_values(d, k_max):
    vals = [d]   # phi(0)
    for k in range(1, k_max + 1):
        prev = vals[-1]
        new = d + (2/d) * (prev**2 - prev**3)
        vals.append(new)
    return np.array(vals)

# values of d
d_values = [0.0001, 0.001]

# choose how many k values you want
k_max = 5
k = np.arange(k_max + 1)

plt.figure(figsize=(8, 5))

for d in d_values:
    vals = phi_values(d, k_max)
    plt.plot(k, vals, marker='o', label=f'd = {d}')

plt.xlabel('k')
plt.ylabel(r'$\phi(k)$')
plt.title(r'$\phi(k)$ for different values of $d$')
plt.legend()
plt.grid(True)
plt.show()
```
